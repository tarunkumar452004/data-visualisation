```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/content/most_runs_average_strikerate.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
sns.pairplot(df)
```

```python
a=df[df['numberofballs']>2000]
a
```

```python
a[a['average']==a['average'].max()]
```

```python
a[a['numberofballs']==a['numberofballs'].max()]
```

```python
a[a['strikerate']==a['strikerate'].max()]
```

```python
a[a['total_runs']==a['total_runs'].max()]
```

```python
f=a.sort_values(by ='strikerate',ascending = False).head(10)
f
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5), dpi=150)
plt.bar(f['batsman'],f['strikerate'])
plt.xlabel("batsman")
plt.ylabel("strikerate")
plt.title('Top 10 Strike rates')
plt.legend(["STRIKERATES"])
plt.xticks(rotation=90)
plt.show()
```

```python
b=a.sort_values(by ='total_runs',ascending = False).head(10)
b
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5), dpi=150)
plt.bar(b['batsman'],b['total_runs'])
plt.xlabel("batsman")
plt.ylabel("total_runs")
plt.title('Top 10 most runs')
plt.legend(["runs"])
plt.xticks(rotation=90)
plt.show()
```
